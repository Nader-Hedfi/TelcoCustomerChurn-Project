```python
# ============================================================
# üåç Customer Concentration in the state of California
# ============================================================

# Aggregate customer data by city
city_concentration = location_data.groupby('city').agg({
    'customer_id': 'count',       # Count the number of customers per city
    'latitude': 'mean',           # Average latitude for the city
    'longitude': 'mean',          # Average longitude for the city
}).reset_index()

city_concentration.rename(columns={'customer_id': 'customer_count'}, inplace=True)

# Converting the DataFrame to a GeoDataFrame
geometry = [Point(xy) for xy in zip(city_concentration['longitude'], city_concentration['latitude'])]
geo_city_data = gpd.GeoDataFrame(city_concentration, geometry=geometry, crs="EPSG:4326")

# Loading the shapefile from Natural Earth
shapefile_path = "../ne_110m_admin_1_states_provinces/ne_110m_admin_1_states_provinces.shp"
states = gpd.read_file(shapefile_path)

# Filter for California
california = states[states['name'] == 'California']

# Ensure geometries are valid
california = california[california.is_valid]
geo_city_data = geo_city_data[geo_city_data.is_valid]

# Align CRS
geo_city_data = geo_city_data.to_crs(california.crs)


# Get the top 5 cities by customer count
biggest_cities = geo_city_data.nlargest(5, 'customer_count')

# Plot the customer concentration
plt.figure(figsize=(12, 8))
base = california.plot(color='lightgray', edgecolor='black', figsize=(12, 8))

# Overlay the customer data
geo_city_data.plot(
    ax=base,
    markersize=geo_city_data['customer_count'] * 10, 
    color='cornflowerblue',
    alpha=0.7
)

# Annotate the biggest cities
for _, row in biggest_cities.iterrows():
    plt.text(
        row['geometry'].x, row['geometry'].y,
        row['city'], fontsize=13, ha='right', color='m'
    )

plt.title("Customer Concentration by City in California", fontsize=16)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
# ============================================================
# üõç  Product distribution among Customers 
# ============================================================

# Adjusting the Yes and No to 1 and 0 in the Dataframe
for column in online_services.columns[1:-1]:  # Exclude 'Customer_ID' and 'Internet_Type'
    online_services[column] = online_services[column].apply(lambda x: 1 if x == "Yes" else 0)

product_columns = ["phone_service", "internet_service", "online_security", "online_backup", 
                   "device_protection", "premium_tech_support", "streaming_tv", 
                   "streaming_movies", "streaming_music"]

# Count the number of '1's (Yes) for each column
yes_counts = online_services[product_columns].sum()

# Creating the bar chart
plt.figure(figsize=(10, 6))
ax = yes_counts.plot(kind='bar', color='blue', alpha=0.7)

# Adding labels and title
plt.title("Number of Customers Enrolled in Product/Service Type", fontsize=14)
plt.xlabel("Products/Services", fontsize=12)
plt.ylabel("Count of 'Yes' (1)", fontsize=12)
plt.xticks(rotation=45, ha='right')  
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numbers on top of the bars
for i, count in enumerate(yes_counts):
    plt.text(i, count + 2, str(count), ha='center', va='bottom', fontsize=10)

# plt.tight_layout()  
plt.show()
```

```python
# ============================================================
# üõç Multi Product distribution among Customers 
# ============================================================

product_columns = ["phone_service", "internet_service", "online_security", "online_backup", 
                   "device_protection", "premium_tech_support", "streaming_tv", 
                   "streaming_movies", "streaming_music"]

# Calculating the total number of products each client is enrolled in
online_services['total_products'] = online_services[product_columns].sum(axis=1)
product_counts = online_services['total_products'].value_counts().sort_index()

# Plotting the data
plt.figure(figsize=(10, 6))
ax = product_counts.plot(kind='bar', color='blue', alpha=0.7)

# Adding labels and title
plt.title("Distinct Number of Clients by Number of Enrolled Products", fontsize=14)
plt.xlabel("Number of Products Enrolled", fontsize=12)
plt.ylabel("Number of Clients", fontsize=12)
plt.xticks(rotation=0)

# Annotate the bars with counts
for i, count in enumerate(product_counts):
    plt.text(i, count + 5, str(count), ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)
ax.grid(False)
plt.tight_layout()
plt.show()
```

```python
###########################################################
# Analyzing Churn Rates by Payment Method and Contract Type
###########################################################

# Encode categorical variables
df = pd.get_dummies(payment_analysis, columns=['payment_method', 'contract', 'paperless_billing'], drop_first=True)
# Dropping non relevant categorical columns
df = df.drop(columns=['customer_id', 'customer_status', 'churn_label', 'churn_category', 'churn_reason'])

# Define independent variables (features) and dependent variable (target)
X = df.drop(columns=['churn_value'])  # Drop target column
y = df['churn_value']  # Target column

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fitting Logistic Regression Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

## Model Evaluation
# Predictions
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]  # Probabilities for the positive class (churn=1)

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("Confusion Matrix:")
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# AUC-ROC Curve
roc_auc = roc_auc_score(y_test, y_pred_proba)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='grey')
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

## Model Coefficients
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
}).sort_values(by='Coefficient', ascending=False)
print("Logistic Regression Coefficients:")
print(coefficients)
```

### &#x1F4CC;	Key Findings

The coefficients indicate the strength and direction of the relationship between each feature and the likelihood of churn, where positive values increase churn probability and negative values decrease it.

**Top Negative Contributors to Churn:**
